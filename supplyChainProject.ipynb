```python
def dataframe_check(df):
  '''
  fungsi ini akan mencari 
  kolom null, data duplicated, persentasi missing,
  unik value, data type dan mengembalikannya ke dalam bentuk
  dataframe
  '''
  missing = df.isnull().sum()
  duplicate = df.duplicated().sum()
  missing_per = round(missing/len(df),4)*100
  unique_val = df.nunique()
  type_data = df.dtypes
  sample_unique = [df[col].unique() for col in df.columns]
  df = pd.DataFrame({'Missing_values':missing,
                    'Percent of Missing (%)':missing_per,
                    'Duplicated Data':duplicate,
                    'Numbers of Unique':unique_val,
                    'Data type':type_data,
                    'Sample Unique':sample_unique}).sort_values("Percent of Missing (%)",ascending=False)
  return df
```

#### **1. Mengidentifikasi Missing Values pada Dataset**


```python
plt.style.use('ggplot')
missing = df.isnull().sum()
missing_per = round(missing/len(df) * 100,3)
# membuat dataframe untuk kolom missing value
missing_df = pd.DataFrame({
    'Total Missing Rows' : missing,
    'Percentage' :missing_per
}).sort_values('Total Missing Rows',ascending=False)

# hanya menampilkan kolom yang memiliki missing value
missing_df = missing_df[missing_df['Total Missing Rows'] > 0].reset_index()


missing_df
```

```python
# membuat fungsi barplot 

def barplot_annot(data,x,y,title,xlabel,ylabel):
    fig, ax = plt.subplots(figsize = (10,3))

    ax = sns.barplot(
        data=data,
        x=x,
        y=y,
        hue=x,
        palette='viridis'
    )

    ax.set(
        title = title,
        xlabel = xlabel,
        ylabel = ylabel
    )

    for bar in ax.patches:
        ax.annotate(f'{bar.get_height():.0f}',  
                    (bar.get_x() + bar.get_width() / 2., bar.get_height()),  
                    ha='center', va='center',  # Posisi teks anotasi (horizontal dan vertikal)
                    xytext=(0,-7),  # Jarak teks anotasi dari bar (horizontal, vertikal)
                    textcoords='offset points',  # Koordinat teks anotasi
                    fontsize=9,  # Ukuran font teks anotasi
                    color='white',  # Warna teks anotasi
                    fontweight='semibold'  # Ketebalan huruf teks anotasi
                    )

    plt.show()

barplot_annot(missing_df,'index','Total Missing Rows','Total Missing Value', 'Column','Total')
barplot_annot(missing_df,'index','Percentage','Percentage Missing Value', 'Column','Percent')
```

#### **D. Asosiasi Kolom Kategorikal**

```python
from scipy.stats import chi2_contingency
pd.set_option('display.max_colwidth',None)
pd.set_option('display.max_rows',None)


def chi2_test(df, column1, column2, alpha=0.05):
    """
    Lakukan uji chi-square independensi antara dua kolom kategorikal dalam DataFrame.

    Args:
        df (DataFrame): DataFrame yang berisi data.
        column1 (str): Nama kolom pertama.
        column2 (str): Nama kolom kedua.
        alpha (float): Tingkat signifikansi yang digunakan (default=0.05).

    Returns:
        str: Hasil uji chi-square.
    """
    contingency_table = pd.crosstab(df[column1], df[column2])
    stat, p_value, dof, expected_freq = chi2_contingency(contingency_table)

    # result = f"Nilai uji statistik Chi squared: {stat:.4f}\n"
    result = f"P-value: {p_value:.4f} "

    if p_value < alpha:
        result += "Terdapat keterkaitan antara kedua kolom."
    else:
        result += "Tidak terdapat keterkaitan antara kedua kolom."

    return result



# Daftar nama kolom yang akan diuji
kolom_uji = ['Type', 'Delivery Status', 'Category Name', 'Customer City',
       'Customer Country', 'Customer Segment', 'Customer State',
       'Customer Street', 'Department Name', 'Market', 'Order City',
       'Order Country', 'Order Region', 'Order State', 'Order Status',
       'Product Name', 'Shipping Mode']

# Inisialisasi list untuk menyimpan hasil uji chi-square
hasil_uji = []

# Melakukan uji chi-square untuk setiap pasangan kolom
for i in range(len(kolom_uji)):
    for j in range(i+1, len(kolom_uji)):
        result = chi2_test(df, kolom_uji[i], kolom_uji[j])
        hasil_uji.append({'Kolom 1': kolom_uji[i], 'Kolom 2': kolom_uji[j], 'Hasil Uji': result})

# Membuat DataFrame dari hasil uji
df_hasil_uji = pd.DataFrame(hasil_uji)

df_hasil_uji

```

```python
import matplotlib.cm as cm

def plot_bar_pie_with_annotation(data, x, y, hue, title, xlabel, ylabel,suptitle): 
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

    # Plot barplot on the first column
    ax1 = sns.barplot(
        data=data,  # dataframe yang akan di visualisasi
        x=x,   # kolom yang menjadi sumbu x
        y=y,    # kolom yang menjadi sumbu y
        hue=hue,
        palette='viridis',  # di buat berdasarkan gradasi warna viridis
        ax=ax1
    )  

    ax1.set(
        title=title,
        xlabel=xlabel,
        ylabel=ylabel
    )

    for bar in ax1.patches:
        ax1.annotate(f'{bar.get_height():.0f}',
                    (bar.get_x() + bar.get_width() / 2., bar.get_height()),
                    ha='center', va='center',  # Posisi teks anotasi (horizontal dan vertikal)
                    xytext=(0, -7),  # Jarak teks anotasi dari bar (horizontal, vertikal)
                    textcoords='offset points',  # Koordinat teks anotasi
                    fontsize=9,  # Ukuran font teks anotasi
                    color='white',  # Warna teks anotasi
                    fontweight='semibold'  # Ketebalan huruf teks anotasi
                    )

    ax1.tick_params(axis='x', rotation=45)  # Memutar label sumbu x pada barplot

    #color viridis 
    viridis = cm.viridis([0.4, 0.6])

    # Plot pie chart on the second column

    ax2.pie(data[y], labels=data[x], autopct='%1.1f%%', startangle=90, textprops={'color': 'white'},
    colors = viridis , explode = [0.1,0])

    ax2.set(title='Percentage')  # Set title untuk pie plot

    centre_circle = plt.Circle((0, 0), 0.4, fc='white')
    fig.gca().add_artist(centre_circle)
    # Set overall title
    plt.suptitle(suptitle, fontsize=16, fontweight='normal')

    plt.tight_layout()
    plt.show()
```

```python
delivery_status = df_late.groupby(['Delivery Status','Late_delivery_risk'])[['Late_delivery_risk']].count().rename(
    columns={'Late_delivery_risk':'Total'}
).reset_index()

plot_bar_pie_with_annotation(
    data=delivery_status,
    x='Delivery Status',
    y='Total',
    title='Delivery Status',
    hue='Delivery Status',
    xlabel='Delivery Status', 
    ylabel='Total Shipping',
    suptitle='Total Delivery Status (Late)'
)
```

```python
def annotate_bars_dollar(ax):
    '''fungsi untuk memberikan anotasi pada bar'''
    for bar in ax.patches:
        x, y = bar.get_xy()
        ax.text(
            x + bar.get_width(), y + bar.get_height()/2, f'$ {bar.get_width():.2f}',
            va='center', ha='right', color='white',fontsize=10,fontweight='semibold'
        )
    return ax

def annotate_bars(ax):
    '''fungsi untuk memberikan anotasi pada bar'''
    for bar in ax.patches:
        x, y = bar.get_xy()
        ax.text(
            x + bar.get_width(), y + bar.get_height()/2, f'{bar.get_width():.0f}',
            va='center', ha='right', color='white',fontsize=10,fontweight='semibold'
        )
    return ax

def plot_barh_total_median(data,y1,x1,hue,y2,x2,xlabel1,xlabel2,title1,title2,suptitle,label):
    fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2, figsize=(14,6)) 
    
    ax1 = sns.barplot(
        data = data,
        y = y1,
        x = x1,
        orient='h',
        hue = hue,
        palette = 'viridis',
        ax=ax1
    )
    
    ax1.set(
        title = title1,
        xlabel = xlabel1,
        ylabel = label
    )
    
    ax2 = sns.barplot(
        data = data,
        y = y2,
        x = x2,
        orient='h',
        hue = hue,
        palette = 'viridis',
        ax=ax2
    )
    ax2.set(
        title = title2,
        xlabel = xlabel2,
        ylabel = label
    )
    
    annotate_bars(ax1)
    annotate_bars_dollar(ax2)
    
    plt.suptitle(suptitle, fontsize=16, fontweight='normal')
    plt.tight_layout()
    plt.show()
```

```python
def annotate_bars_dollar(ax):
    '''fungsi untuk memberikan anotasi pada bar'''
    for bar in ax.patches:
        x, y = bar.get_xy()
        ax.text(
            x + bar.get_width(), y + bar.get_height()/2, f'$ {bar.get_width():.2f}',
            va='center', ha='right', color='white',fontsize=10,fontweight='semibold'
        )
    return ax

def annotate_bars(ax):
    '''fungsi untuk memberikan anotasi pada bar'''
    for bar in ax.patches:
        x, y = bar.get_xy()
        ax.text(
            x + bar.get_width(), y + bar.get_height()/2, f'{bar.get_width():.0f}',
            va='center', ha='right', color='white',fontsize=10,fontweight='semibold'
        )
    return ax

def plot_bar_total_median(data,y1,x1,hue,y2,x2,xlabel1,xlabel2,title1,title2,suptitle):
    fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1, figsize=(14,12)) 
    
    ax1 = sns.barplot(
        data = data,
        y = y1,
        x = x1,
        hue = hue,
        palette = 'viridis',
        ax=ax1
    )
    
    ax1.set(
        title = title1,
        xlabel = xlabel1,
        ylabel = 'Product Name'
    )
    
    ax2 = sns.barplot(
        data = data,
        y = y2,
        x = x2,
        hue = hue,
        palette = 'viridis',
        ax=ax2
    )
    ax2.set(
        title = title2,
        xlabel = xlabel2,
        ylabel = 'Product Name'
    )
    
    annotate_bars(ax1)
    annotate_bars_dollar(ax2)
    
    plt.suptitle(suptitle, fontsize=16, fontweight='normal')
    plt.tight_layout()
    plt.show()
```
